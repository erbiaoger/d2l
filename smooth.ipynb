```python
import numpy as np
import matplotlib.pyplot as plt

import smoothn

xp = np.arange(0, 1, 0.02)

x,y = np.meshgrid(xp, xp)
f = np.exp(x+y) + np.sin((x-2*y)*3)
fn = f + np.random.randn(*f.shape)*0.1

fs = smoothn.smoothn(fn)

```

```python
fn.shape
```

```python
fs[0].shape
```

```python
%matplotlib inline

fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(121)
ax1.imshow(fn)
ax1.set_title('Noisy')
ax2 = fig.add_subplot(122)
ax2.imshow(fs[0])
ax2.set_title('Smoothed')
```

```python
import scipy.io as sio

data = sio.loadmat('hour10_hv.mat')['hour10']
```

```python
data.shape
```

```python
data_smooth = smoothn.smoothn(data, smoothOrder=0.1)
```

```python
%matplotlib inline

fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(131)
ax1.imshow(data.T, aspect='auto')
ax1.set_title('Noisy')
ax2 = fig.add_subplot(132)
ax2.imshow(data_smooth[0].T, aspect='auto')
ax2.set_title('Smoothed')
ax3 = fig.add_subplot(133)
ax3.imshow(data_smooth[0].T-data.T, aspect='auto')
```

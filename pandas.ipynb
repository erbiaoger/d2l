## 二、常用的列切片操作

```python
# 1、取一列的所有数据
index = 0
column_name = 'sepal_length'

df[column_name]         ##利用列名直接取某一列
df[df.columns[index]]   ##适合于你不知道列名，但是知道它在第几列
```

```python
# 2、根据位置取行列数据df.loc
index = 0

df.loc[index]                                    ##根据行的位置，取特定行数据（列全取）
df.loc[[index],['sepal_length', 'sepal_width']]  ##取index行的，ab两列数据
df.loc[[index],'sepal_length':'petal_width']     ##取index行的，列名为'a' 到 列名为 'b'的所有列
##总之，列不能通过索引来取数
```

```python
# 3、根据索引位置来取数df.iloc

df.iloc[0:10,0:10] ##切片后面的值取不到，即col_index=10,row_indx=10取不到
df.iloc[[0,5,10],[1, 3, 4]] ##可按照需求，选择特定的行和列
##总之iloc之内的数据都是数字，不能是行名列名
```

```python
# 4、根据条件，逻辑值索引取数

df[df.A>0] ##取出A列中大于0的数，
df[df['A'].isin(['one','two'])] ##取出A列中包含'one','two'的数据,这个功能很强大，##可以帮助我们filter出符合条件的数据。
```

```python
# 5、给列赋值

df['A']=np.array([1]*len(df))   ##用数组给某列赋值
df.loc[:,['a','c']]=[]          ##根据位置赋值
##知道如何取数，就能轻松给数据框赋值啦。
```

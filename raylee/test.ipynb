```python
from functools import partial

class Model:
    def __init__(self):
        self.vsv = 1
        self.vpv = 2
        self.rhov = 3
        self.h = 4
        self.fks = 5
        self.Nn = 6
        
model = Model()

def Raylee_Forward(vsv, vpv, rhov, h, fks, Nn):
    # 在这里实现 Raylee_Forward 函数
    result = vsv + vpv + rhov + h + fks + Nn
    return result

# 创建一个部分应用的函数
def Forw(vsv, vpv, rhov, h, fks, Nn):
    # 在这里调用 Raylee_Forward 函数，并传递参数
    return Raylee_Forward(vsv, vpv, rhov, h, fks, Nn)

# 使用 functools.partial 创建一个新函数，其中一些参数已经固定
Forw_with_model = partial(Forw, model.vpv, model.vsv, model.h, model.fks, model.Nn)

# 调用新函数并传递 vsv 参数
some_vsv_value = 1
result = Forw_with_model(some_vsv_value)
```
